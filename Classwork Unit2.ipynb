collect() - collect common powers of a term in an expression.

```python
from sympy import *
x, y, z = symbols('x, y, z')
eqn = x*y + x - 3 + 2*x**2 - z*x**2 + x**3
collected_eqn = collect(eqn, x)
collected_eqn
```

```python
collected_eqn.coeff(x)
```

Q. Write a program which finds roots of the quadratic equations $x^2 + x + 1$.

```python
import sympy as sp
x = sp.Symbol('x')
eqn = x**2 + x + 1
roots = sp.solve(eqn,x)
roots
```

Q. Write a program which finds roots of the cubic equation $x^3 - 6x^2 + 11x - 6$

```python
import sympy as sp
x = sp.symbols('x')
equation = x**3 - 6*x**2 + 11*x - 6
roots = sp.solve(equation,x)
print("Roots:", roots)
```

Q. Write a program which finds roots of any Quadratic equation.

```python
import cmath

def solve_quad():
    a = int(input("Enter the coefficient for x**2:"))
    b = int(input("Enter the coefficient for x:"))
    c = int(input("Enter the constant term:"))
    eqn = a*x**2 + b*x + c
    print("Equation:",eqn)
    root1 = (-b + sqrt(b**2 - 4*a*c))/(2*a)
    root2 = (-b - sqrt(b**2 - 4*a*c))/(2*a)
    print("\nRoots are", f"{root1} and {root2}")

solve_quad()
```

```python
import cmath

def find_roots(a,b,c):
    val = sqrt(b**2 - 4*a*c)
    root1 = (-b + val) / (2*a)
    root2 = (-b - val) / (2*a)

    return root1, root2

a = int(input("Enter the coefficient a: "))
b = int(input("Enter the coefficient b: "))
c = int(input("Enter the coefficient c: "))
roots = find_roots(a,b,c)
print("Root 1:", roots[0])
print("Root 2:", roots[1])
```

Q. WAPP for the given quadratic eqn and list of numbers and display the numbers that satify the quadratic eqn.

```python
import sympy as sp

x = sp.symbols('x')

def input_eqn():
    a = int(input("Enter the coefficient for x^2:"))
    b = int(input("Enter the coefficient for x:"))
    c = int(input("Enter the constant term:"))
    eqn = a*x**2 + b*x + c
    return eqn

def input_list():
    n = int(input("\nEnter the no. of numbers you want to check for:"))
    num_list = []
    for i in range(n):
        num = int(input(f"Enter the number {i+1}:"))
        num_list.append(num)
    return num_list
        
def sol_check(eqn,num_list):
    sol_list = []
    for num in num_list:
        if eqn.subs(x,num) == 0:
            sol_list.append(num)
    return sol_list
        
eqn = input_eqn()
print("\nEquation:", eqn)
num_list = input_list()
print("\nSatisfying numbers:",sol_check(eqn,num_list))
```
