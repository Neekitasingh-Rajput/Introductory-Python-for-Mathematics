```python
#Use a for loop to print an upside-down triangle like the one below. Allow the user to specify how high the triangle should be.
'''
****
***
**
*
'''

def upside_down_triangle(n):
    for i in range(n,0,-1):
        for j in range(i):
            print("*",end='')
        print()

n = int(input("How high should the triangle be?:"))           
upside_down_triangle(n)
```

```python
#Reverse a list.

def reverse_list():
    list = []
    n = int(input("Enter the no of elements:"))
    for i in range(1,n+1):
        a = input(f" Enter the element {i}")
        list.append(a)
    print("List:", list)
    print("Reversed List:", list[::-1])

reverse_list()
```

```python
#Write a program to add two lists index-wise. Create a new list that contains the 0th index item from both the list, 
#then the 1st index item, and so on till the last element. Any leftover items will get added at the end of the new list.

list_1 = [1,2,3,4,5,6,7,8]
list_2 = ['a','b','c','d']

concatenated_list = []
max_length =max(len(list_1), len(list_2))
    
for i in range(max_length):
    item1 = str(list_1[i]) if i < len(list_1) else ''
    item2 = str(list_2[i]) if i < len(list_2) else ''
    concatenated_list.append(item1 + item2)

print("List_1:", list_1)
print("List_2:", list_2)
print("Concatenated_list:", concatenated_list)
```

```python
#Write a Python program that defines the quadratic equation 2x^2+3x+1=0. The program should:
'''
● Take a list of numbers as input from the user.
● Check whether these numbers satisfy the equation 2x^2+3x+1=0 or not.
● Print the numbers that satisfies the equation.
'''

def satisfies_equation(x):
    return 2*x**2 + 3*x + 1 == 0
        
n = int(input("Enter the number of elements:"))

numbers = []
satisfying_numbers = []

for i in range(n):
    num = int(input(f"Enter the number[{i+1}]:"))
    numbers.append(num)
    if satisfies_equation(num):
        satisfying_numbers.append(num)

print("\nThe list of numbers:", numbers)
print("The numbers that satisfies the equation:", satisfying_numbers)
```

```python
#Write a Python program that takes a list of numbers as input from the user and then 
#performs the following operations:
'''
● Sort the list in ascending order.
● Find the maximum and minimum values from the list.
● Remove any duplicate values from the list.
● Print the final modified list.
'''
def input_list():
    my_list = []
    n = int(input("Enter the number of elements:"))
    for i in range(n):
        num = int(input(f"Enter the element [{i+1}]"))
        my_list.append(num)
    return my_list

my_list = input_list()

print("\nList:", my_list)
print("Sorted list:", sorted(my_list))
print("Maximum value:", max(my_list))
print("Minimum value:", min(my_list))
print("Final modified list:", sorted(set(my_list)))
```

```python
#3. Write a Python program that allows a user to input student names and their scores. 
'''
Store the data in a dictionary where the key is the student's name and the value is their score. 
The program should:
● Allow the user to input a new student and score.
● Display the student with the highest score.
● Allow the user to search for a specific student and display their score.
'''

def input_dict():
    my_dict = {}
    s = int(input("Enter the number of students:"))
    for i in range(s):
        name = input(f"Enter the name of student[{i+1}]:")
        score = int(input(f"Enter the score for {name}:"))
        my_dict[name] = score
    return my_dict

def add_new_student(student_dict):
    n = int(input("\nEnter the number of new students to add: "))
    for i in range(n):
        name = input(f"Enter the name of new student[{i+1}]: ")
        score = int(input(f"Enter the score for {name}: "))
        student_dict[name] = score
    return student_dict

def highest_score(student_dict):
    highest_score_student = max(student_dict, key=student_dict.get)
    return highest_score_student, student_dict[highest_score_student]

def search_student(student_dict):
    name = input("\nEnter the name of the student to search for: ")
    if name in student_dict:
        print(f"{name}'s score: {student_dict[name]}")
    else:
        print(f"{name} not found in the student dictionary.")

student_dict = input_dict()
print("\nStudents_dict:", student_dict)

new_student_dict = add_new_student(student_dict)
print("\nStudents dict updated:", new_student_dict)

highest_student, highest_score_value = highest_score(new_student_dict)
print(f"\nThe student with the highest score is {highest_student} with a score of {highest_score_value}.")

search_student(new_student_dict)
```

```python
#Q2. Write a Python program that generates and prints multiplication tables for a range of 
#numbers from start to end. The user should input the starting and ending numbers, and the 
#program should print the multiplication tables for each number in this range from 1 to 10.

def mul_tables(start,end):
    for i in range(start,end+1):
        print(f"\nMultiplication table for {i} is:")
        for k in range(1,11):
            print(f"{i}x{k} ={i*k}")

start = int(input("Enter the starting number:"))
end = int(input("Enter the ending number:"))
mul_tables(start,end)
```

```python
#Q4. Consider a doctor's office where each appointment is 30 minutes long. 
#Write a Python program that checks if a given time slot is available for 
#an appointment within a specified range of hours.

def doc_appointments(slot_start, slot_end):
    slots = []
    for hour in range(slot_start, slot_end + 1):
        for minute in [0, 30]:
            slots.append(f"{hour:02d}:{minute:02d}")
    return slots

def check_slot_availability(slot_start, slot_end, slot_to_check):
    available_slots = doc_appointments(slot_start, slot_end)
    if slot_to_check in available_slots:
        print(f"\nYes, the slot {slot_to_check} is available!")
    else:
        print(f"\nNo, the slot {slot_to_check} is not available.")

slot_start = int(input("Enter the starting hour (24-hour format): "))
slot_end = int(input("Enter the ending hour (24-hour format): "))
slot_to_check = input("Enter the slot to check (HH:MM): ")

check_slot_availability(slot_start, slot_end, slot_to_check)
```

```python
"""
Q5. Develop a simple ATM machine simulation in Python. The ATM should allow the user to perform
basic operations such as checking the balance, depositing money, withdrawing money, and exiting
the program.
Requirements:
i. Initial Setup:  Start with an initial balance of $1000. 
ii. Menu Options: Provide a menu with the following options: Check Balance Deposit Money
Withdraw Money Exit Functionality:
iii. Check Balance: Display the current balance. 
iv. Deposit Money: Prompt the user to enter the amount
to deposit. Ensure the amount is a positive number.
v. Withdraw Money: Prompt the user to enter the amount to withdraw. Ensure the amount is a
positive number and that sufficient funds are available. 
vi. Exit: Terminate the program with a thank you
message.
vii. Error Handling: Handle invalid inputs such as non-numeric values and ensure that the user cannot
withdraw more than the available balance.

For Instance:

 --- ATM Menu ---
1. Check Balance
2. Deposit Money
3. Withdraw Money
4. Exit Select an option (1-4): 3
Enter the amount to withdraw: $600
Withdrew $600.00.
New balance: $400.00

--- ATM Menu ---

1. Check Balance
2. Deposit Money
3. Withdraw Money
4. Exit Select an option (1-4): 4
Thank you for using the ATM. Goodbye!
"""

def ATM():
    balance = 1000.00
    
    def check_balance():
        nonlocal balance
        print(f"\nBalance: {balance:.2f}")
        
    def deposit_money(amt_to_deposit):
        nonlocal balance
        if amt_to_deposit <= 0:
            print("Invalid amount!")
        else:
            balance = balance + amt_to_deposit
            print(f"\n ${amt_to_deposit:.2f} succesfully deposited.")
            print(f" New Balance: {balance:.2f}")
        
    def withdraw_money(amt_to_withdraw):
        nonlocal balance
        if amt_to_withdraw <=0:
            print("Invalid amount!")
        elif amt_to_withdraw > balance:
            print("Insufficient balance!")
        else:
            balance = balance - amt_to_withdraw
            print(f"\n ${amt_to_withdraw:.2f} succesfully withdrawn.")
            print(f" New Balance: {balance:.2f}")

    def exit_program():
        print("\nThank you for using the ATM. Goodbye!")
        
    
    while True:
        print("\n----ATM MENU----")
        print("1. Check Balance \n2. Deposit Money \n3. Withdraw Money \n4. Exit")
        print("----------------")
        
        try:
            a = int(input("Select an option (1-4):"))

            if a == 1:
                check_balance()

            elif a == 2:
                amt_to_deposit = float(input("\nEnter amount to deposit (in $):"))
                deposit_money(amt_to_deposit)

            elif a == 3:
                amt_to_withdraw = float(input("\nEnter amount to withdraw (in $):"))
                withdraw_money(amt_to_withdraw)

            elif a == 4:
                exit_program()
                break

            else:
                print("Invalid selection! Please choose a valid option (1-4).")
                
        except ValueError:
            print("\nInvalid input! Please enter a valid number.")

ATM()
```

```python
#Extract the elements and add it to a list, where elements should not be repeated.

def list_input():
    n = int(input("Enter the no. of elements you want to add:"))
    my_list = []
    for i in range(n):
        element = eval(input(f"Enter the element {i+1}:"))
        my_list.append(element)
    return my_list
    
def list_without_repeatation(my_list):
    print("List:", my_list)
    my_list = list(set(my_list))
    print("List without any repeatation:", my_list)

my_list = list_input()
list_without_repeatation(my_list)
```
