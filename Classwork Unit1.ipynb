```python
#Extract the elements and add it to a list, where elements should not be repeated.

def list_input():
    n = int(input("Enter the no. of elements you want to add:"))
    my_list = []
    for i in range(n):
        element = eval(input(f"Enter the element {i+1}:"))
        my_list.append(element)
    return my_list
    
def list_without_repeatation(my_list):
    print("List:", my_list)
    my_list = list(set(my_list))
    print("List without any repeatation:", my_list)

my_list = list_input()
list_without_repeatation(my_list)
```

```python
#Write a program to find the intersection of sets that contain tuples.

def set_input():
    n = int(input("Enter the no. of elements you want to add:"))
    my_list = []
    for i in range(n):
        element = eval(input(f"Enter the element {i+1}:"))
        my_list.append(element)
    my_set = set(my_list)
    return my_set
    
def set_intersection(set1, set2):
    return set1 & set2
    #use nested loop to check

set1 = set_input()
set2 = set_input()

print("\nSet1:", set1)
print("Set2:", set2)
print("\nSet intersection:", set_intersection(set1, set2))
```
