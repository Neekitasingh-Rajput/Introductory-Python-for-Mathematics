```python
#1.Write a program to find the intersection of sets that contain tuples. (define your own function)

def set_input(set_name):
    n = int(input(f"\nEnter the no. of elements you want to add for {set_name}:"))
    my_list = []
    for i in range(n):
        element = eval(input(f"Enter the element {i+1}:"))
        my_list.append(element)
    my_set = set(my_list)
    return my_set
    
def set_intersection(set1, set2):
    common = set()
    for i in set1:
        for j in set2:
            if i == j:
                common.add(i)
    return common
                

set1 = set_input('set1')
set2 = set_input('set2')

print("\nSet1:", set1)
print("Set2:", set2)
print("\nSet intersection:", set_intersection(set1, set2))
```

```python
#2. Write a program to find the symmetric difference between multiple sets.

def symmetric_difference(setA, setB):
    sym_diff = set()
    for elem in setA:
        if elem not in setB:
            sym_diff.add(elem)
    for elem in setB:
        if elem not in setA:
            sym_diff.add(elem)
    return sym_diff

setA = set_input('Set A')
setB = set_input("Set B")

print("\nSetA:", setA)
print("SetB:", setB)
print("\nSymmetric Difference:", symmetric_difference(setA, setB))
```

```python
#3. Write a program to find the Cartesian product of two sets.

def cartesian_product(set1, set2):
    cart_product = set()
    for i in set1:
        for j in set2:
            ordered_pair = (i,j)
            cart_product.add(ordered_pair)
    return cart_product

my_set1 = set_input('Set A')
my_set2 = set_input('Set B')

print("\nSeta:", my_set1)
print("SetB:", my_set2)
print("\nCartesian Product (AXB):", cartesian_product(my_set1, my_set2))
```

```python
#4. Generate a dictionary where the keys are numbers, and the values are Boolean indicating if the number is prime.

def input_list():
    n = int(input("Enter the no. of numbers you will add:"))
    my_list = []
    for i in range(n):
        elem = int(input(f"Enter the element {i+1}:"))
        my_list.append(elem)
    return my_list

def is_prime(a):
    if a <= 1:
        return False
    else:
        for i in range(2,a):
            if a % i == 0:
                return False
        return True

def prime_check(my_list):
    _dict = {}
    for i in my_list:
        _dict[i] = is_prime(i)
    return _dict


my_list = input_list()
print("\nList of numbers:", my_list)
print("\nPrime check:", prime_check(my_list))
```

```python
#5. Given a list of integers, write a program to compute the frequency of each integer and store the results in a dictionary.

def input_list():
    n = int(input("Enter the no. of numbers you will add:"))
    my_list = []
    for i in range(n):
        elem = int(input(f"Enter the element {i+1}:"))
        my_list.append(elem)
    return my_list
    
def freq(my_list):
    freq_dict = {}
    for i in set(my_list):
        count = my_list.count(i)
        freq_dict[i] = count
    return freq_dict
    
my_list = input_list()
print("\nEntered list:", my_list)
print("Frequency Dictionary:", freq(my_list))
```

```python
#6. Given a list of sentences, count the occurrences of each word across all sentences. 
#Store the counts in a dictionary and print the dictionary.

def input_list_sentences():
    n = int(input("Enter the no. of sentences you will add:"))
    _list = []
    for i in range(n):
        elem = str(input(f"Enter the sentence {i+1}:"))
        _list.append(elem.strip(".,!?"))
    return _list

def count_words(sentences):
    words = []
    for i in sentences:
        for j in i.split():
            words.append(j.lower())
    count_dict = {}
    for k in set(words):
        count = words.count(k)
        count_dict[k] = count
    return count_dict

sentences = input_list_sentences()
print("\n Entered list of sentences:", sentences)
print("\n Count of each word across all the sentences:", count_words(sentences))
```

```python
#7.Given a paragraph (a string), split it into words, count the frequency of each word, and store the counts in a dictionary. 
#Print the dictionary showing each word and its frequency.

import string

def input_para():
    paragraph = str(input("Enter a paragraph:"))
    return paragraph

def word_count(paragraph):
    word_counts = {}
    cleaned_para = paragraph.translate(str.maketrans('', '', string.punctuation)).lower()
    for word in cleaned_para.split():
        word_counts[word] = word_counts.get(word, 0) + 1
    return word_counts

my_para = input_para()
print("Entered paragraph:", my_para)
print("Word count of each word in paragraph:", word_count(my_para))
```

```python
#8. Given a list of integers, group them by their first digit and store the results in a dictionary where the keys 
#are the first digits and the values are lists of numbers starting with that digit. Print the dictionary.

def input_list():
    n = int(input("Enter the number of integers you want to add: "))
    my_list = []
    for i in range(n):
        elem = int(input(f"Enter integer {i+1}: "))
        my_list.append(elem)
    return my_list

def group_by_first_digit(numbers):
    grouped_dict = {}
    for num in numbers:
        first_digit = str(num)[0]
        if first_digit not in grouped_dict:
            grouped_dict[first_digit] = []
        grouped_dict[first_digit].append(num)
    return grouped_dict

numbers = input_list()
print("\nEntered list of numbers:", numbers)
print("\nGrouped by first digit:", group_by_first_digit(numbers))
```
